```python
sns.scatterplot(x="category", y="diff_nutrition",data=df_final);
plt.xlabel("Categories")
plt.title("Distribution of categories by calories")
plt.ylabel('# Calories')
plt.savefig('../project-II-pipelines/images/chart2.png', dpi=100)

plt.show()
```

```python
sns.histplot(x=df_final.diff_nutrition)
plt.xlabel("Calories")
plt.title("Distribution of calories")
plt.ylabel('# Cocktails')
plt.savefig('../project-II-pipelines/images/chart3.png', dpi=100)
plt.show()
```

```python
top_cocktails = df_final.sort_values(by='diff_nutrition', ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_cocktails['name'], top_cocktails['diff_nutrition'])
plt.xlabel('Cocktail Name')
plt.ylabel('Calories')
plt.title('Top 5 Cocktails by Calories')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.savefig('../project-II-pipelines/images/chart4.png', dpi=100)
plt.show()
```

```python
top_cocktails = df_final.sort_values(by='diff_nutrition', ascending=True).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_cocktails['name'], top_cocktails['diff_nutrition'])
plt.xlabel('Cocktail Name')
plt.ylabel('Calories')
plt.title('Top 5 Cocktails with less calories')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.show()
```

```python
glass_counts = df_final['diff_glass'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
glass_counts.plot(kind='bar')
plt.xlabel('Glass Type')
plt.ylabel('Number of Cocktails')
plt.title('Number of Cocktails by Glass Type')
plt.xticks(fontsize=8)
plt.savefig('../project-II-pipelines/images/chart5.png', dpi=100)

plt.show()
```

```python
garnish_counts = df_final['garnish'].value_counts().reset_index().head(10)
garnish_counts.columns = ['Garnish', 'Count']

plt.figure(figsize=(10, 6))
sns.barplot(data=garnish_counts, x='Count',y="Garnish", hue='Garnish', palette='viridis',legend=False)
plt.xlabel('Number of Cocktails')
plt.ylabel('Garnish')
plt.title('Top 10 Used Garnishes in Cocktails')
plt.xticks(rotation=90,fontsize=8)  # Rotate the x-axis labels for better readability
plt.savefig('../project-II-pipelines/images/chart6.png', dpi=100)
plt.show()
```
